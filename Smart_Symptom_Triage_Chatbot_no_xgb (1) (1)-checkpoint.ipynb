## Section 1: Data Description

```python
# Load the preprocessed dataset
import pandas as pd
import os
desktop = r'C:\Users\SAMR TRADERS\OneDrive\Desktop'
os.chdir(desktop)
# Read the file
df = pd.read_csv('Final(processed dataset).csv')
print(df.head())
print(f"Shape: {df.shape}")
# Information of shape
print(f"Dataset Shape : {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"Total Features: {df.shape[1] - 2}  (174 binary symptom features)")
print(f"Target Columns: 'disease' (46 classes) and 'triage_level' (4 classes)")
print(f"Missing Values: {df.isnull().sum().sum()}")
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

### Features of Dataset

```python
# Identify symptom columns
symptom_cols = [c for c in df.columns if c.startswith('symptom_') or c.startswith('rare_symptom_')]
common_symptoms = [c for c in symptom_cols if c.startswith('symptom_')]
rare_symptoms   = [c for c in symptom_cols if c.startswith('rare_symptom_')]

print(f"Common symptoms: {len(common_symptoms)}")
print(f"Rare symptoms: {len(rare_symptoms)}")
print(f"Total features: {len(symptom_cols)}")
print(f"\nTriage Level Distribution:")
print(df['triage_level'].value_counts())
print(f"\nTop 5 Diseases:")
print(df['disease'].value_counts().head())
```

## Section 2: Data Preprocessing

```python
# Missing value check
print("Step 1:-")
print("Missing Values:", df.isnull().sum().sum(), "(none)\n")

# Duplicate check
print("Step 2:-")
print("Duplicates:", df.duplicated().sum(), "(none)\n")

# Feature / Target split
X = df[symptom_cols].values.astype(np.int8)   # binary matrix (0/1)
y_disease = df['disease']
y_triage  = df['triage_level']
print("Step 3:-")
print(f"X shape: {X.shape}  (int8 binary matrix)")
print(f"y_disease unique: {y_disease.nunique()}")
print(f"y_triage  unique: {y_triage.nunique()}")
```

### Label Encoding & Train/Test Split

```python
le_disease = LabelEncoder()
le_triage  = LabelEncoder()
y_disease_enc = le_disease.fit_transform(y_disease)
y_triage_enc  = le_triage.fit_transform(y_triage)

print("Disease classes:", list(le_disease.classes_))
print("Triage classes :", list(le_triage.classes_))

joblib.dump(le_disease, 'models/le_disease.pkl')
joblib.dump(le_triage,  'models/le_triage.pkl')
pickle.dump(symptom_cols, open('models/symptom_cols.pkl', 'wb'))

#Train/Test split : 80/20
X_train, X_test, yd_train, yd_test, yt_train, yt_test = train_test_split(X, y_disease_enc, y_triage_enc,test_size=0.2, random_state=42, stratify=y_triage_enc
)
print(f"\nTraining samples : {X_train.shape[0]:,}")
print(f"Testing  samples : {X_test.shape[0]:,}")
```

## Section 4: Modeling

```python
# Helper function to evaluate and print metrics
def evaluate(y_true, y_pred, name, classes=None):
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    rec  = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1   = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    print("----------------------------")
    print(f"| {name}   |")
    print("----------------------------")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}  (weighted avg)")
    print(f"Recall: {rec:.4f}  (weighted avg)")
    print(f"F1-Score: {f1:.4f}  (weighted avg)")
    if classes is not None:
        print(classification_report(y_true, y_pred, target_names=classes, zero_division=0))
    return {'accuracy':round(acc,4),'precision':round(prec,4),'recall':round(rec,4),'f1':round(f1,4)}
```

### Model A: Random Forest (Triage)

```python
# Baseline RF for triage prediction
rf_triage = RandomForestClassifier(
    n_estimators=100,   # Number of decision trees
    max_depth=15,       
    random_state=42,
    n_jobs=-1           # use all CPU cores
)
rf_triage.fit(X_train, yt_train)
yt_pred = rf_triage.predict(X_test)
m_rf_triage = evaluate(yt_test, yt_pred, "RF — Triage (Baseline)", list(le_triage.classes_))
joblib.dump(rf_triage, 'models/rf_triage.pkl')
print("Saved: models/rf_triage.pkl")
```

### Model B: Random Forest (Disease)

```python
# RF for disease prediction (subset for speed)
idx = np.random.RandomState(42).choice(len(X_train), 30000, replace=False)
rf_disease = RandomForestClassifier(n_estimators=50, max_depth=20, random_state=42, n_jobs=-1)
rf_disease.fit(X_train[idx], yd_train[idx])
yd_pred = rf_disease.predict(X_test)
m_rf_disease = evaluate(yd_test, yd_pred, "RF: Disease (Baseline)", list(le_disease.classes_))
joblib.dump(rf_disease, 'models/rf_disease.pkl')
print("Saved: models/rf_disease.pkl")
```

### Model C: Logistic Regression (Baseline)

```python
# LR as a simple baseline for comparison
scaler = StandardScaler()
Xtr_s = scaler.fit_transform(X_train)
Xte_s = scaler.transform(X_test)
lr = LogisticRegression(max_iter=500, random_state=42)
lr.fit(Xtr_s, yt_train)
ylr_pred = lr.predict(Xte_s)
m_lr = evaluate(yt_test, ylr_pred, "Logistic Regression: Triage", list(le_triage.classes_))
joblib.dump(lr,'models/lr_triage.pkl')
joblib.dump(scaler, 'models/scaler.pkl')
```

## Section 6: Regression (MAE, RMSE, R²)

```python
# Encode triage as numeric (0-3) for regression
le_reg = LabelEncoder()
y_num  = le_reg.fit_transform(df['triage_level']).astype('float32')
print("Numeric mapping:", dict(zip(le_reg.classes_, le_reg.transform(le_reg.classes_))))

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X, y_num, test_size=0.2, random_state=42)

idx_r = np.random.RandomState(42).choice(len(Xr_train), 20000, replace=False)
reg = RandomForestRegressor(n_estimators=50, max_depth=15, random_state=42, n_jobs=-1)
reg.fit(Xr_train[idx_r], yr_train[idx_r])
yr_pred = reg.predict(Xr_test)

mae  = mean_absolute_error(yr_test, yr_pred)
rmse = np.sqrt(mean_squared_error(yr_test, yr_pred))
r2   = r2_score(yr_test, yr_pred)

print(f"\nRegression Metrics:")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")
```

## Section 7: Clustering (Silhouette Score)

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

idx_c = np.random.RandomState(0).choice(len(X), 10000, replace=False)
Xc = X[idx_c]

# Matching 4 triage levels
km = MiniBatchKMeans(n_clusters=4, random_state=42, batch_size=2048, n_init=5)
km.fit(Xc)
cluster_labels = km.predict(Xc)

sil = silhouette_score(Xc, cluster_labels, sample_size=3000, random_state=42)
print(f"KMeans (k=4) Silhouette Score: {sil:.4f}")
print("Range: -1 (worst) to +1 (best)")
print("Low score is expected — diseases share many overlapping symptoms")
```

### Cluster Visualization (PCA)

```python
# Project 174 dimensions down to 2D using PCA for visualization
pca = PCA(n_components=2, random_state=42)
Xc_2d = pca.fit_transform(Xc[:2000])
lbl_2d = cluster_labels[:2000]

fig, ax = plt.subplots(figsize=(7, 5))
for k, c in zip(range(4), ['#e74c3c','#3498db','#2ecc71','#9b59b6']):
    mask = lbl_2d == k
    ax.scatter(Xc_2d[mask,0], Xc_2d[mask,1], s=12, alpha=0.5, color=c, label=f'Cluster {k+1}')
ax.set_title(f'KMeans (k=4) — PCA 2D View\nSilhouette Score = {sil:.4f}', fontweight='bold')
ax.set_xlabel('PC1'); ax.set_ylabel('PC2')
ax.legend(); ax.set_facecolor('#f8f9fa')
fig.tight_layout()
plt.savefig('plots/09_clustering_pca.png', dpi=120, bbox_inches='tight')
plt.show()
```

## Section 9: Improved Models

```python
from sklearn.neural_network import MLPClassifier
rf_tuned = RandomForestClassifier(
    n_estimators=150,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42, n_jobs=-1
)
rf_tuned.fit(X_train, yt_train)  
m_rf_tuned = evaluate(yt_test, rf_tuned.predict(X_test), "Tuned RF — Triage", list(le_triage.classes_))
joblib.dump(rf_tuned, 'models/rf_triage_tuned.pkl')
print("Saved: models/rf_triage_tuned.pkl")
```

### Neural Network (MLP)

```python
# NEW: Neural Network — 2 hidden layers (128, 64 neurons)
nn_scaler = StandardScaler()
Xtr_scaled = nn_scaler.fit_transform(X_train)
Xte_scaled = nn_scaler.transform(X_test)
joblib.dump(nn_scaler, 'models/nn_scaler.pkl')

mlp_triage = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    max_iter=20,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=3,
    learning_rate_init=0.001,
    random_state=42
)
mlp_triage.fit(Xtr_scaled[:40000], yt_train[:40000])
print(f"Neural Network stopped at epoch: {mlp_triage.n_iter_}")
m_nn = evaluate(yt_test, mlp_triage.predict(Xte_scaled), "Neural Network — Triage", list(le_triage.classes_))
joblib.dump(mlp_triage, 'models/mlp_triage.pkl')
```
